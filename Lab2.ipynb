```python
import tkinter
from tkinter import *
from tkinter import filedialog

from PIL import ImageTk, Image
```

```python
class Train:
    
    def __init__(self, y, window_width):
        self.window_width = window_width
        self.x = window_width * 3 / 4
        self.y = y
        self.is_running = False
        self.is_active = True
        
        self.head_image_data = Image.open("B:\\Code\\PythonLabs\\Resource\\train_head.png").convert("RGBA")
        self.head_image = ImageTk.PhotoImage(self.head_image_data)
        self.segment_image_data = Image.open("B:\\Code\\PythonLabs\\Resource\\train_segment.png").convert("RGBA")
        self.segment_image = ImageTk.PhotoImage(self.segment_image_data)
        
    def start(self):
        self.is_running = True
        
    def stop(self):
        self.is_active = False
    
    def update(self):
        if self.is_running:
            self.x += 8
            
    def draw(self, canvas):
        if self.is_active:
            canvas.create_image(self.x, self.y, image = self.head_image)
            
            temp = self.x - 180
            while temp > 0:
                canvas.create_image(temp, self.y, image = self.segment_image)
                temp -= 218
            canvas.create_image(temp, self.y, image = self.segment_image)
```

```python
class TicketCollector:
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.is_active = True
        
        self.image_data = Image.open("B:\\Code\\PythonLabs\\Resource\\ticket_collector.png").convert("RGBA")
        self.image = ImageTk.PhotoImage(self.image_data)
        
    def stop(self):
        self.is_active = False
            
    def draw(self, canvas):
        if self.is_active:
            canvas.create_image(self.x, self.y, image = self.image)
```

```python
class Person:
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.is_active = True
        
        self.image_data = Image.open("B:\\Code\\PythonLabs\\Resource\\person.png").convert("RGBA")
        self.image = ImageTk.PhotoImage(self.image_data)
        
    def move(self, dx, dy):
        self.x += dx
        self.y += dy
        
    def stop(self):
        self.is_active = False
            
    def draw(self, canvas):
        if self.is_active:
            canvas.create_image(self.x, self.y, image = self.image)
```

```python
class Scene:
    
    def __init__(self):
        self.window_width = 1024
        self.window_height = 720
        
        self.root = Tk()
        self.root.geometry(f'{self.window_width}x{self.window_height}')
        self.root.bind('<Up>', self.up_event)
        self.root.bind('<Down>', self.down_event)
        self.root.bind('<Left>', self.left_event)
        self.root.bind('<Right>', self.right_event)
        
        self.widget = Canvas(self.root, bg = 'white')
        self.widget.pack(fill = BOTH, expand = 1)
        
        self.train = Train(180, self.window_width)
        self.ticket_collector = TicketCollector(150, 150)
        self.person = Person(self.window_width - 100, self.window_height - 100)
        
    def check_event(self):
        if self.ticket_collector.is_active:
            if self.person.x < 512 and self.person.y < 512:
                self.ticket_collector.stop()
        elif self.person.is_active and self.person.y < 300:
            self.person.stop()
            self.train.start()
        if self.ticket_collector.is_active and self.person.is_active and self.person.y < 300:
            self.train.stop()
        
    def up_event(self, event):
        self.person.move(0, -10)
        self.check_event()
        
    def down_event(self, event):
        self.person.move(0, +10)
        self.check_event()
        
    def left_event(self, event):
        self.person.move(-10, 0)
        self.check_event()
        
    def right_event(self, event):
        self.person.move(+10, 0)
        self.check_event()
    
    def update_event(self):
        self.train.update()
        
        self.widget.delete(ALL)
        self.train.draw(self.widget)
        self.ticket_collector.draw(self.widget)
        self.person.draw(self.widget)
        
        self.widget.after(10, self.update_event)
        
    def loop(self):
        self.update_event()
        self.widget.mainloop()
```

```python
Scene().loop()
```
