```python
import math
import numpy
import matplotlib.pyplot as pyplot
import pandas

import seaborn
palette = seaborn.light_palette("#c06dc5", as_cmap = True)

%matplotlib inline
```

```python
def table(function, interval_from = -1, interval_to = 1, step = 0.2, size = 20, horisontal = False):
    x = numpy.arange(interval_from, interval_to, step)
    y = [function(xi) for xi in x]
    data = pandas.DataFrame(data = y, index = x, columns = ["tan(x) / x"])
    if horisontal:
        data = data.transpose()
    return data.head(int((interval_to - interval_from) / step)).style.background_gradient(cmap = palette, axis = 0)
    # print(data)
```

```python
def graph(function, interval_from = -2, interval_to = 2, step = 0.01, size = 20, limit_bottom = -10, limit_top = +10):
    x = numpy.arange(interval_from, interval_to, step)
    y = [function(xi) for xi in x]
    temp = pyplot.figure(figsize = (size, size))
    seaborn.lineplot(x, y)
```

```python
f = lambda x: math.tan(x) / x
```

```python
table(f, interval_from = -1, interval_to = 1.1, step = 0.2, horisontal = False)
```

```python
graph(f, size = 10, interval_from = -1, interval_to = +1, step = 0.01)
```

```python
graph(f, size = 10, interval_from = -10, interval_to = +10, step = 0.02)
```
